```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('f:/data/sonar.csv', header=None)
df
```

```python
dataset=df.values
x=dataset[:,0:60].astype(float)
y=dataset[:,60]
```

```python
y
le=LabelEncoder()
le.fit(y)
y=le.transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    x, y, test_size=0.3, random_state=1025)
```

```python
#모델 설정
model=Sequential()
model.add(Dense(32, input_dim=60, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy'])
```

```python
# 모델 생성
model.fit(xtrain, ytrain, epochs=100, batch_size=5)
```

```python
model.evaluate(xtest, ytest)
```

```python
model.save('my_model.h5') # 로컬에 모델 저장
```

```python
n_fold=10
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skf=StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=1025)
```

```python
x.shape #(208, 60)
y.shape #(208,)
```

```python
acc=[]

#모델 설정, 컴파일, 실행
for train, test in skf.split(x,y):
    
    #모델 설정
    model=Sequential()
    model.add(Dense(32, input_dim=60, activation='relu'))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    # 모델 컴파일
    model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy'])

    # 모델 생성
    model.fit(xtrain, ytrain, epochs=100, batch_size=5)
    imsi_acc=model.evaluate(x[test],y[test])[1]
    acc.append(imsi_acc)

```

```python
print(acc)
print(sum(acc)/10.0)
```

```python
from sklearn.datasets import load_iris
```

```python
fold_iris = load_iris()
```

```python
fold_iris.keys()
xdata=fold_iris.data
ydata=fold_iris.target
```

```python
from sklearn.model_selection import KFold
```

```python
kfold=KFold(n_splits=5)
```

```python
cv_acc=[]
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
```

```python
from sklearn.metrics import accuracy_score
```

```python
for train_idx, test_idx in kfold.split(xdata):
    # print(xdata[train_idx].shape) #(120, 4)
    # print(xdata[test_idx].shape) #(30, 4)
    xtrain, xtest = xdata[train_idx], xdata[test_idx]
    # print(ydata[train_idx].shape) #(120,)
    # print(ydata[test_idx].shape) #(30,)
    ytrain, ytest = ydata[train_idx], ydata[test_idx]
    #의사결정트리 모델 생성
    clf.fit(xtrain,ytrain)
    #예측
    pred=clf.predict(xtest)

    #정확도 
    acc=accuracy_score(ytest, pred)
    cv_acc.append(acc)
```

```python
print(cv_acc)
print(np.mean(cv_acc))
```

```python
from keras.datasets import mnist
```

```python
(xtrain, ytrain), (xtest, ytest)=mnist.load_data()
```

```python
xtrain.shape #(60000, 28, 28)
ytrain.shape #(60000,)
xtest.shape #(10000, 28, 28)
ytest.shape #(10000,)
```

```python
# 훈련 데이터와 검증 데이터 분리
xval=xtrain[50000:]
yval=ytrain[50000:]
xtrain=xtrain[:50000]
ytrain=ytrain[:50000]
```

```python
xtrain = xtrain.reshape(50000, 784).astype('float32') / 255.0
xval = xval.reshape(10000, 784).astype('float32') / 255.0
xtest = xtest.reshape(10000, 784).astype('float32') / 255.0
```

```python
train_indx=np.random.choice(50000, 700)
val_indx=np.random.choice(10000,300)
```

```python
xtrain=xtrain[train_indx]
ytrain=ytrain[train_indx]
xval=xval[val_indx]
yval=yval[val_indx]
```

```python
from keras.utils import np_utils
```

```python
ytrain=np_utils.to_categorical(ytrain)
yval=np_utils.to_categorical(yval)
ytest=np_utils.to_categorical(ytest)
```

```python
model=Sequential()
model.add(Dense(32, input_dim=28*28, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
#콜백 함수 : 사전에 정의한 조건을 충족하게 되면, 시스템에 의해 자동으로 호출되는 함수
```

```python
es=EarlyStopping(patience=30)
```

```python
hist=model.fit(xtrain, ytrain, epochs=100, 
               batch_size=10, validation_data=(xval, yval), callbacks=[es])
```
