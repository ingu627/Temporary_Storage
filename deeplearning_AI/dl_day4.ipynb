```python
# 3ì°¨ì› : pca1 ~ pca3
pca.explained_variance_
# ê° ì¶•ì´ ì„¤ëª…í•˜ëŠ” ë°ì´í„°ì˜ í‘œí˜„ë ¥, ê°’ì´ í´ìˆ˜ë¡ ì„¤ëª…ì„ ë§Žì´ í•˜ê³  ìžˆëŠ” ì¶•
# ì²«ë²ˆì§¸ ì¶•ì€ ì•½ 3ì„ ê°–ê³ , ì„¸ë²ˆì§¸ ì¶• 0.14ì´ë¯€ë¡œ 
# ì²«ë²ˆì§¸ pca1ì¶•ì´ ì„¸ë²ˆì§¸ pca3ì¶•ë³´ë‹¤ ì„¤ëª…ë ¥ì´ ì•½ 20ë°° ì´ìƒì´ë‹¤.

# ì›ëž˜ ë°ì´í„° 
# x:ì‹ ìž¥, y:ëª¸ë¬´ê²Œ, z:ë¨¸ë¦¬ì¹´ë½ê¸¸ì´

# array([2.93808505, 0.9201649 , 0.14774182])
# ex) ë‚¨ë…€ë¥¼ ì‹ë³„(3ê°€ì§€ íŠ¹ì„±)
# ì™¸í˜•(PCA1) : 2.93808505 (73%)
# ëª©ì†Œë¦¬(PCA2) : 0.9201649 (22.8%)
# ì´ë¦„(PCA3) : 0.14774182 (3%)
```
