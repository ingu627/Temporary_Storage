```python
df3 = pd.DataFrame({
    'A': [1, 3, 4, 3, 4],
    'B': [2, 3, 1, 2, 3],
    'C': [1, 5, 2, 4, 4]
})
df3
#df3의 각 열에 대해 최대값을 출력하시오.
df3.max() -df3.min() # 길이가 동일하기 때문에 가능

#df3의 각 행에 대해 최대값을 출력하시오
df3.max(axis=1)
```

```python
# cut : 구간의 경계값 지정 
# qcut: 개수가 동일한 여러 구간으로 나누고자 할 때 사용 
```

```python
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]
```

```python
mycut=pd.cut(ages, [0, 20, 30, 50, 70, 100], labels=['미성년자', '청년','중년', '장년','노년'])
#개구간(소괄호, 포함x) / 폐구간(대괄호, 포함o)
# (0.0, 20.0] = 0 초과 20 이하
```

```python
mycut.categories # categories : 카테고리의 종류를 알 수 있다.
```

```python
mycut.codes 
# NaN은 -1로 표현됨
```

```python
agedf = pd.DataFrame(ages, columns=['ages'])
bin=[0,20,30,50,70,100]
label=['미성년자','청년', '중년', '장년', '노년']

agedf['age_cut'] = pd.cut(ages, bins=bin, labels=label)
agedf
```

```python
# to_list() : 리스트로 바꿔주는 함수
```

```python
# age_cut 은 문자열이 아닌 카테고리 타입

agedf

#comb열 추가 : NaNO, 미성년자2, 미성년자10
agedf['comb'] = agedf.age_cut.astype(str) + agedf.ages.astype(str)
agedf

# astype : 타입을 지정해줌
# dytpe : 타입을 확인함

```
