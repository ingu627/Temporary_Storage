```python
dataset=np.loadtxt('f:/data/pima-indians-diabetes.csv', delimiter=',')
```

```python
xvalid = xtrain[500:, :8]
xvalid.shape
```

```python
yvalid=ytrain[500:]
yvalid.shape
```

```python
xtrain=xtrain[0:500, :8]
ytrain=ytrain[0:500]
```

```python
# model.fit()ì„¤ì •
# xtrain, ytrain ë°ì´í„°ë¡œ í•™ìŠµ -> xvalid, yvalidë¡œ ê²€ì¦ , metrics=accuracy
# => ëª¨ë¸ íŠœë‹
# => xtest, ytest ë°ì´í„°ë¡œ í…ŒìŠ¤íŠ¸ ì§„í–‰ -> ì •í™•ë„ ì¶œë ¥
```

```python
# ëª¨ë¸ì„¤ì •
model=Sequential()

model.add(Dense(64, input_dim=8, activation='relu'))

model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# ëª¨ë¸ ìµœì í™” í•¨ìˆ˜ ë“± ì •ì˜
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

hist=model.fit(
    xtrain, ytrain,
    epochs=600,
    batch_size=2,
    validation_data=(xvalid, yvalid), verbose=1)

print(max(hist.history['val_accuracy']))
print(max(hist.history['accuracy']))
```

```python
from sklearn.preprocessing import StandardScaler

xvalid=StandardScaler().fit_transform(xvalid)
xtrain=StandardScaler().fit_transform(xtrain)
model=Sequential()
model.add(Dense(64, input_dim=8, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
hist = model.fit(xtrain, ytrain, epochs = 300, batch_size=10, validation_data=(xvalid,yvalid))
plt.plot(hist.history['accuracy'])
```

```python
fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()
loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')
acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')
```
